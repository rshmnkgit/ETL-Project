### Extract the tution fees for top rated national universities

```python
# Import web scraing dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# Actiate the chromebrowser
executable_path = {"executable_path": "chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# Open the webpage
url = "https://www.usnews.com/best-colleges/rankings/national-universities"
browser.visit(url) 
```

```python
html = browser.html
soup = bs(html, 'html.parser')
```

```python
import time
for x in range(2):    
    html = browser.html
    # Parse HTML with Beautiful Soup
    soup = bs(html, 'html.parser')
    time.sleep(2)
    print("PLEASE WAIT....... Processing ..................")
    try:
        load = browser.find_by_xpath("//button[contains(@class,'Button-suns4m-1 LoadMoreWrapper__LoadMore-zwyk5c-1 bFRTND')]")
        load.first.click()
    except:
        pass
print(".......Processing Completed..........")
```

```python
college_name = []
college_loc = []
univ_rank = []
tution_out = []
tution_in = []
enrollment = []

html = browser.html
# Parse HTML with Beautiful Soup
soup = bs(html, 'html.parser')
# Retrieve all elements that contain book information
all_containers = soup.find_all('div', class_='Box-w0dun1-0 DetailCardColleges__CardContainer-cecerc-0 hZXiKO detail-card-colleges')

for container in all_containers:
    box_college = container.find('div',class_='Box-w0dun1-0 DetailCardColleges__CardOverview-cecerc-5 dBfnVc')    
    anchor = box_college.find('a', class_='Anchor-byh49a-0 DetailCardColleges__StyledAnchor-cecerc-6 dJzybH card-name')
    col_name = anchor.text    
    col_loc = box_college.find('p', class_='Paragraph-sc-1iyax29-0 bZPbig').text    
    NU_Rank = box_college.find('div', class_='RankList__Rank-sc-2xewen-2 kDCAuc ranked')    
    box_tution = container.find('div', class_='Box-w0dun1-0 DetailCardColleges__CardStats-cecerc-10 gvEOiR')    
    get_data = box_tution.find_all('dd', class_='QuickStatHug__Description-hb1bl8-1 iDVTlX')    
    if len(get_data)==2:
        tout_fee = get_data[0].text    
        tin_fee = 'N/A'
        enroll = get_data[1].text
    elif len(get_data)==3:
        tout_fee = get_data[0].text
        tout_fee = tout_fee.replace('(out-of-state)','')
        tin_fee = get_data[1].text
        tin_fee = tin_fee.replace('(in-state)','')
        enroll = get_data[2].text 

    college_name.append(col_name)
    college_loc.append(col_loc)
    univ_rank.append(NU_Rank.text)
    tution_out.append(tout_fee)
    tution_in.append(tin_fee)
    enrollment.append(enroll)

print(f"{college_name}, {college_loc}, {univ_rank}, {tution_out}, {tution_in}, {enrollment}")
```

```python
college_df = pd.DataFrame({'College Name':college_name, 
                           'College Loc':college_loc, 
                           'Rank':univ_rank, 
                           'Tution Out':tution_out, 
                           'Tution In':tution_in, 
                           'Enrollment':enrollment})
college_df
```

```python
browser.quit()
```

### Transform

```python
college_df['Rank'] = college_df['Rank'].str.strip('#')
college_df['Tution Out'] = college_df['Tution Out'].str.strip('$')
college_df['Tution In'] = college_df['Tution In'].str.strip('$')
college_df
```

```python
college_df['Tution In'].replace(",", "", regex=True, inplace=True)
college_df['Tution Out'].replace(",", "", regex=True, inplace=True)
college_df['Enrollment'].replace(",", "", regex=True, inplace=True)
college_df.head()
```

### Load the data into SQLite

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine('sqlite:///Output/College_db.sqlite')
conn = engine.connect()
Base = declarative_base()
Base.metadata.create_all(engine)
```

```python
college_df.to_sql('toprated_colleges', con=conn, if_exists='replace')
```
